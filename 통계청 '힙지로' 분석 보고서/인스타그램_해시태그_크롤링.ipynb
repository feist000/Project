## 3) 크롤링 시작
- 좋아요 40 이상일 경우, 해시태그 가져오기 or 40 이하일 경우, 본문 내용 가져오기 
- 다음 페이지 자동 클릭 (할당 된 범위까지 자동으로 넘어가면서 크롤링)

```python
import re
from bs4 import BeautifulSoup
import unicodedata   # 추가, 2020.08.20

like_tag = []
def get_content(driver):
    # ① 현재 페이지 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    # ② 본문 내용 가져오기
    try:
        content = soup.select('div.C4VMK > span')[0].text
        content = unicodedata.normalize('NFC', content)   # 신규 추가. 2020.08.20
    except:
        content = ' '
    # ③ 본문 내용에서 해시태그 가져오기(정규식 활용)
    tags = re.findall(r'#[^\s#,\\]+', content)  
    # ④ 작성일자 정보 가져오기
    
    # ⑤ 좋아요 수 가져오기
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1]
        like = int(like)
    except:
        like = 0
        
    data = [like, tags]
    if like > 40 :
        if len(data[1]) != 0 :
             like_tag.append(data)
        else :
            data[1] = content
            like_tag.append(data)
        
    return like_tag

for i in range(500) :
    print(get_content(driver))
    driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()
    time.sleep(2)

    
```
